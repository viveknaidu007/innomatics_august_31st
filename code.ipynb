```python
import requests
from bs4 import BeautifulSoup

def get_imdb_reviews(imdb_id):
    url = f"https://www.imdb.com/title/tt{imdb_id}/reviews"
    headers = {'User-Agent': 'Mozilla/5.0'}
    
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        print(f"Failed to retrieve reviews for IMDb ID: {imdb_id}")
        return None
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find reviews - you may need to adjust the selectors based on the IMDb page structure
    reviews = []
    for review in soup.find_all('span', {'class': 'sc-16ede01-2 gkDSSP'}):
        reviews.append(review.get_text())
    
    return reviews

# Example usage for one movie
example_imdb_id = popular_movies_with_links['imdbId'].iloc[0]
reviews = get_imdb_reviews(example_imdb_id)

print(f"Reviews for IMDb ID {example_imdb_id}:")
print(reviews)
```

```python
import requests
import numpy as np
from bs4 import BeautifulSoup

def scrapper(imdbId):
    id = str(int(imdbId))
    n_zeroes = 7 - len(id)
    new_id = "0"*n_zeroes + id
    URL = f"https://www.imdb.com/title/tt{new_id}/"
    request_header = {'Content-Type': 'text/html; charset=UTF-8', 
                      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0', 
                      'Accept-Encoding': 'gzip, deflate, br'}
    response = requests.get(URL, headers=request_header)  # Use requests.get to fetch the page
    soup = BeautifulSoup(response.text, 'html.parser')     # Use BeautifulSoup to parse the page
    imdb_rating = soup.find('span', attrs={'itemprop': 'ratingValue'})  # Find the rating element
    return imdb_rating.text if imdb_rating else np.nan
```

```python
import pandas as pd

# Load the necessary data
df = pd.read_csv('movies.csv')         # Movie details
df2 = pd.read_csv('ratings.csv')       # User ratings
df3 = pd.read_csv('links.csv')         # IMDb links

# Group user ratings
ratings_agg = df2.groupby('movieId')['rating'].agg(['count', 'mean']).reset_index()
ratings_agg.columns = ['movieId', 'rating_count', 'average_rating']

# Merge with movie information
merged_df = pd.merge(df, ratings_agg, on='movieId')

# Filter movies with more than 50 ratings
popular_movies = merged_df[merged_df['rating_count'] > 50]

# Merge with IMDb links
popular_movies_with_links = pd.merge(popular_movies, df3, on='movieId')
```

```python
sci_fi_movies = popular_movies_with_links[popular_movies_with_links['genres'].str.contains('Sci-Fi')]
```

```python
import numpy as np

def scrapper(imdbId):
    id = str(int(imdbId))
    n_zeroes = 7 - len(id)
    new_id = "0"*n_zeroes + id
    URL = f"https://www.imdb.com/title/tt{new_id}/"
    request_header = {'Content-Type': 'text/html; charset=UTF-8', 
                      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0', 
                      'Accept-Encoding': 'gzip, deflate, br'}
    response = requests.get(URL, headers=request_header)
    soup = BeautifulSoup(response.text, 'html.parser')
    imdb_rating = soup.find('span', attrs={'itemprop': 'ratingValue'})
    return imdb_rating.text if imdb_rating else np.nan

# Fetch IMDb ratings for Sci-Fi movies
sci_fi_movies['imdb_rating'] = sci_fi_movies['imdbId'].apply(scrapper)
```

```python
# Convert IMDb ratings to numeric (some might be 'NaN')
sci_fi_movies['imdb_rating'] = pd.to_numeric(sci_fi_movies['imdb_rating'], errors='coerce')

# Find the movie with the highest IMDb rating
highest_rated_sci_fi = sci_fi_movies.loc[sci_fi_movies['imdb_rating'].idxmax()]

# Get the movieId of the highest rated Sci-Fi movie
highest_rated_movie_id = highest_rated_sci_fi['movieId']
print(f"The movieId of the Sci-Fi movie with the highest IMDb rating is: {highest_rated_movie_id}")
```
